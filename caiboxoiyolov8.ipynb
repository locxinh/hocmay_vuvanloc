```python
from ultralytics import YOLO
# Load a model
data_path = '/kaggle/input/datasetcaiboxoi1/data.yaml'
#'đường dẫn đến file data.yaml trong dataset'
model = YOLO("yolov8n.pt")  # build a new model from YAML

# Train the model
results = model.train(data=data_path, epochs=1000, imgsz=640, device = 0) 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('runs/detect/train/results.csv')
print(df.columns.tolist())  # kiểm tra cột

plt.figure(figsize=(15, 10))

# 1. Loss
plt.subplot(2, 2, 1)
plt.plot(df.index, df['train/box_loss'], label='Train Box Loss')
plt.plot(df.index, df['train/cls_loss'], label='Train Class Loss')
plt.plot(df.index, df['val/box_loss'], label='Val Box Loss')
plt.plot(df.index, df['val/cls_loss'], label='Val Class Loss')
plt.title("Loss per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

# 2. Precision
plt.subplot(2, 2, 2)
plt.plot(df.index, df['precision'], label='Precision')
plt.title("Validation Precision per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Precision")
plt.legend()

# 3. Recall
plt.subplot(2, 2, 3)
plt.plot(df.index, df['recall'], label='Recall')
plt.title("Validation Recall per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Recall")
plt.legend()

# 4. mAP
plt.subplot(2, 2, 4)
plt.plot(df.index, df['mAP50'], label='mAP@0.5')
plt.plot(df.index, df['mAP50-95'], label='mAP@0.5:0.95')
plt.title("Validation mAP per Epoch")
plt.xlabel("Epoch")
plt.ylabel("mAP")
plt.legend()

plt.tight_layout()
plt.show()
```
