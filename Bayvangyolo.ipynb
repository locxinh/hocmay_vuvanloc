```python
from ultralytics import YOLO
# Load a model
epochs = 500
batch_size = 16
lr = 0.001
imgsz = 640
optimizer = 'Adam'
data = '/kaggle/input/bayvangkhe/data.yaml'
#'đường dẫn đến file data.yaml trong dataset'
# Train the model
model = YOLO("yolov8s.pt")
results = model.train(data=data, batch=batch_size, epochs=epochs, imgsz=imgsz, optimizer=optimizer)
```
